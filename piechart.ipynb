```python
import matplotlib.pyplot as plt
%matplotlib inline
exp_vals= [1400,600,410,250]
exp_labels = ["home rent", "Food", "Phonebill", "other utility"]
plt.pie(exp_vals, labels= exp_labels)
```

```python
plt.pie(exp_vals, labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, shadow=True,autopct='%1.7f%%',radius=1.5)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, shadow=True,autopct='%1.7f%%',radius=1.5,explode=[0,0,0.1,0.2])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, shadow=True,autopct='%1.7f%%',radius=1.5,explode=[0,0,0.1,0.2],counterclock=True, startangle=45)
plt.show()
```
